```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Import our data into pandas from excel 

nfl_statistics_path = 'Resources/nfl_statistics.xlsx'
nfl_statistics_df = pd.read_excel(nfl_statistics_path)

nfl_statistics_df
```

```python
# Drop Columns

nfl_statistics_df.drop(["game_id", "gameday", "weekday", "gametime", "location", "total", 
            "old_game_id", "gsis", "nfl_detail_id", "pfr", "pff", "espn", "away_rest", 
            "home_rest",  "away_moneyline", "home_moneyline", "spread_line", 
            "away_spread_odds", "home_spread_odds", "total_line", "under_odds", "over_odds",
            "temp", "wind", "away_qb_id", "home_qb_id", "away_coach", "home_coach", "referee",
            "stadium_id"], axis = 1, inplace=True)

nfl_statistics_df
```

```python
# Drop rows prior to 2011

nfl_statistics_df.drop(nfl_statistics_df.loc[nfl_statistics_df["season"] < 2011].index, inplace=True)                                                             
                                                             
nfl_statistics_df
```

```python
# Combine records of renamed teams

nfl_statistics_df["away_team"].replace({"LA":"LAR/STL", "STL":"LAR/STL", "LV": "LV/OAK", 
            "OAK": "LV/OAK", "SD":"LAC/SD", "LAC":"LAC/SD"},inplace=True)
nfl_statistics_df["home_team"].replace({"LA":"LAR/STL", "STL":"LAR/STL", "LV": "LV/OAK", 
            "OAK": "LV/OAK", "SD":"LAC/SD", "LAC":"LAC/SD"},inplace=True)
nfl_statistics_df.tail(32)
```

```python
# Result column isn't showing on table above so show top 5 results seperately

nfl_statistics_df.result.head(5)
```

```python
# use np select to fill in create new winning/losing columns

# create a list of winning conditions
winning_conditions = [
    (nfl_statistics_df["result"] > 0),
    (nfl_statistics_df["result"] < 0),
    (nfl_statistics_df["result"] == 0)
]

# create a list of losing conditions
losing_conditions = [
    (nfl_statistics_df["result"] < 0),
    (nfl_statistics_df["result"] > 0),
    (nfl_statistics_df["result"] == 0)
]

# Create a list of the values we want to assign for each condition
values = [nfl_statistics_df["home_team"], nfl_statistics_df["away_team"], "tie"]

# create a new column and use np.select to assign values for winning column
nfl_statistics_df["winner"] = np.select(winning_conditions, values)

# create a new columnn and use np.select to assign values for losing column
nfl_statistics_df["loser"] = np.select(losing_conditions, values)

# display updated DataFrame
nfl_statistics_df
```

```python
# export to csv

nfl_statistics_path = 'Resources/nfl_statistics_r1.csv'
nfl_statistics_df.to_csv(nfl_statistics_path)
```
