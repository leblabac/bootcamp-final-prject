```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder 
enc = OneHotEncoder()
```

```python
# Load the final merged dataset
OHE_to_load = "NFL_Merged_df.csv"
OHE_df = pd.read_csv(OHE_to_load)
OHE_df
```

```python

# drop unneeded columns
OHE_df = OHE_df.drop(columns=["Player","Playoff","Won_Superbowl","Win%"])
OHE_df
```

```python
# Using get_dummies and a for loop one hot encode each categorical column and delete the original data.

categorical_columns = ['Pos', 'Cat_POS', 'Team','Conf', 'Division']
for column in categorical_columns:
    tempdf = pd.get_dummies(OHE_df[column], prefix=column)
    OHE_df = pd.merge(
        left=OHE_df,
        right=tempdf,
        left_index=True,
        right_index=True,
    )
    OHE_df = OHE_df.drop(columns=column)
OHE_df
```

```python
NFL_OHE_df_path = "NFL_OHE_df.csv"
OHE_df.to_csv(NFL_OHE_df_path, index = False)
```
