```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder 
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from pathlib import Path
from collections import Counter
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score,mean_squared_error
```

```python
# Load the final merged dataset
OHE_to_load = "NFL_Merged_df.csv"
OHE_df = pd.read_csv(OHE_to_load)
OHE_df
```

```python

# drop unneeded columns
OHE_df = OHE_df.drop(columns=["Player"])
OHE_df
```

```python
# Using get_dummies and a for loop one hot encode each categorical column and delete the original data.

categorical_columns = ['Pos', 'Cat_POS', 'Team','Conf', 'Division']
for column in categorical_columns:
    tempdf = pd.get_dummies(OHE_df[column], prefix=column)
    OHE_df = pd.merge(
        left=OHE_df,
        right=tempdf,
        left_index=True,
        right_index=True,
    )
    OHE_df = OHE_df.drop(columns=column)
OHE_df
```

```python
# write df to new csv removing the index

NFL_OHE_df_path = "NFL_OHE_df.csv"
OHE_df.to_csv(NFL_OHE_df_path, index = False)
```

```python
X = OHE_df.drop(columns=["Playoff","Won_Superbowl","Win%"])

# X= df.loc(columns=["Playoff","Won_Superbowl","Win%"])
# X.head()

y = OHE_df[["Playoff"]]
y2 = OHE_df[["Win%"]]
y3 = OHE_df[["Won_Superbowl"]]
# print(X)
print(y2)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
X_train, X_test, y_train, y_test = train_test_split(X, y2, random_state=78)
# X_train, X_test, y_train, y_test = train_test_split(X, y3, random_state=78)
```

```python
# Creating the scaler instance
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler()
```

```python
# Fitting the scaler
X_scaler = data_scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
# NFL_TTS = pd.DataFrame(NFL_TTS,columns=NFL_Scale.columns)
# NFL_TTS
```

```python
# Create a random forest classifier.
# rf_model = RandomForestClassifier(n_estimators=128, random_state=78)
rf_model = RandomForestRegressor(n_estimators=128, random_state=78)

# Fitting the model
# rf_model = rf_model.fit(X_train_scaled, y_train_encoded)
rf_model = rf_model.fit(X_train_scaled, y_train)
# history = rf_model.fit(X_train_scaled, y_train)
# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)
# print(f" Random forest predictive accuracy: {round(accuracy_score(y_test_encoded,y_pred),3)}")
print(f" Random forest regressor MSE: {round(mean_squared_error(y_test,y_pred),3)}")
```
