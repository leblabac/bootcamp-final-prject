```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Import our data into pandas from CSV
nfl_standings_path = 'Resources/nfl_standings.csv'
nfl_standings_df = pd.read_csv(nfl_standings_path)

nfl_standings_df
```

```python
# Change column headings
nfl_standings_df = nfl_standings_df.rename(columns={"season":"year", "pct":"win %"})
nfl_standings_df
```

```python
# Combine records of renamed teams
nfl_standings_df["team"].replace({"LA":"LAR/STL", "STL":"LAR/STL", "LV": "LV/OAK", 
            "OAK": "LV/OAK", "SD":"LAC/SD", "LAC":"LAC/SD"},inplace=True)
nfl_standings_df.tail(32)
```

```python
# export to csv
nfl_standings_path = 'Resources/nfl_standings_r1.csv'
nfl_standings_df.to_csv(nfl_standings_path)
```
