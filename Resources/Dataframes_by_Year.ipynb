## One Hot Encoding 

```python
# Using get_dummies and a for loop one hot encode each categorical column and delete the original data - 2011

categorical_columns = ['Pos', 'Cat_POS', 'Team',
                       'Conf', 'Division', 'Playoff', 'Won Superbowl']
for column in categorical_columns:
    tempdf = pd.get_dummies(df2011[column], prefix=column)
    df2011 = pd.merge(
        left=df2011,
        right=tempdf,
        left_index=True,
        right_index=True,
    )
    df2011 = df2011.drop(columns=column)
df2011

```
