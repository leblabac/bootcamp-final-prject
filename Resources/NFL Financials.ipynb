```python
# Add the Pandas dependency.
import pandas as pd
```

```python
# Files to load
financials_to_load = "NFL_Yearly_Breakdown_combined.csv"
```

```python
# Read the financials data file and store it in a Pandas DataFrame.
financials_df = pd.read_csv(financials_to_load)
financials_df
```

```python
# Determine if there are any missing values in the financial data.
financials_df.count()
```

```python
# Determine if there are any missing values in the financial data.
financials_df.isnull()
```

```python
# Drop 
financials_df = financials_df.drop(columns=["Rank","Unnamed: 19","Age", "FA Year", "Sign Year", "Sign Age","Cash","Draft Year","AAV","CONT YR","CONT VALUE","Earnings"])
financials_df
```

```python
#define values
values = [2022, 2023, 2024]

#drop rows that contain any value in the list
financials_df = financials_df[financials_df.Year.isin(values) == False]
financials_df
```

```python
# Determine if there are any missing values in the financial data.
financials_df.isnull()
```

```python
financials_df.dtypes
```

```python
# Combine records of renamed teams
financials_df["Team"].replace({"LAR":"LAR/STL", "STL":"LAR/STL", "JAC":"JAX","LV": "LV/OAK","OAK": "LV/OAK", "SD":"LAC/SD", "LAC":"LAC/SD"},inplace=True)
financials_df.head(50)
```

```python
# export to csv
financials_df_path = 'financials_r1.csv'
financials_df.to_csv(financials_df_path)
```
