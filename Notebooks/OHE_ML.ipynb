```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder 
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from pathlib import Path
from collections import Counter
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score,mean_squared_error
```

```python
# Load the final merged dataset
OHE_to_load = "../Resources/NFL_Merged_df.csv"
OHE_df = pd.read_csv(OHE_to_load)
OHE_df
```

```python

# drop unneeded columns

OHE_df_limited = OHE_df.drop(columns=["Player","Wins","Losses","Ties","Team","Cat_POS","Conf","Division","Div_Rank","Scored","Allowed","Net","SOV","SOS"])

OHE_df_limited
```

```python
# OHE Using get_dummies and a for loop one hot encode categorical column and delete the original data.

categorical_columns = ['Pos']
for column in categorical_columns:
    tempdf = pd.get_dummies(OHE_df_limited[column], prefix=column)
    OHE_df_limited = pd.merge(
        left=OHE_df_limited,
        right=tempdf,
        left_index=True,
        right_index=True,
    )
    OHE_df_limited = OHE_df_limited.drop(columns=column)
OHE_df_limited
```

Begin machine learning code:

```python
#set up the X and y values

X = OHE_df_limited.drop(columns=["Playoff","Won_Superbowl","Win%"])
y = OHE_df_limited[["Playoff"]]

# train,test, split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

print(X_train.shape)
print(X_test.shape)
```

```python
# Creating the scaler instance
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler()
# Fitting the scaler
X_scaler = data_scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# only for categorical target to convert Yes\NO to 0/1
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
LE = label_encoder.fit(y_train)
y_train_encoded = LE.transform(y_train)
y_test_encoded = LE.transform(y_test)
print(y_train_encoded)
```

```python
# Playoff as the target/output/solving for
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78)

# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train_encoded)

# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)
print(f"Playofs as Target- Random forest predictive accuracy: {round(accuracy_score(y_test_encoded,y_pred),3)}")
```

```python
# Won_Superbowl as the target/output/solving for

# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78)

# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train_encoded)

# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)
print(f"Won_Superbowl target- Random forest predictive accuracy: {round(accuracy_score(y_test_encoded,y_pred),3)}")
```

```python
# Win% as the target/output/solving for errors
# Create a random forest Regressor.
rf_model = RandomForestRegressor(n_estimators=128, random_state=78)

# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)

# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)

# Print out the results
print(f" Win% Random forest regressor MSE: {round(mean_squared_error(y_test,y_pred),3)}")
```
