```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# reading csv files
nfl_standings = pd.read_csv('..Resources/nfl_standings_r5.csv')
nfl_financials = pd.read_csv('..Resources/financials_r1.csv')
```

```python
# Merging datasets
NFL_Merged_df = pd.merge(nfl_financials, nfl_standings,  how='left', left_on=['Year','Team'], right_on = ['Year','Team'])
NFL_Merged_df
```

```python
NFL_Merged_df = NFL_Merged_df.drop(columns="Unnamed: 0_x")
```

```python
NFL_Merged_df = NFL_Merged_df.drop(columns="Unnamed: 0_y")
```

```python
Positions = [
    (NFL_Merged_df["Pos"] == "C"),
    (NFL_Merged_df["Pos"] == "FB"),
    (NFL_Merged_df["Pos"] == "G"),
    (NFL_Merged_df["Pos"] == "LT"),
    (NFL_Merged_df["Pos"] == "QB"),
    (NFL_Merged_df["Pos"] == "RB"),
    (NFL_Merged_df["Pos"] == "RT"),
    (NFL_Merged_df["Pos"] == "T"),
    (NFL_Merged_df["Pos"] == "TE"),
    (NFL_Merged_df["Pos"] == "WR"),
    (NFL_Merged_df["Pos"] == "CB"),
    (NFL_Merged_df["Pos"] == "DE"),
    (NFL_Merged_df["Pos"] == "DT"),
    (NFL_Merged_df["Pos"] == "FS"),
    (NFL_Merged_df["Pos"] == "ILB"),
    (NFL_Merged_df["Pos"] == "LB"),
    (NFL_Merged_df["Pos"] == "OLB"),
    (NFL_Merged_df["Pos"] == "S"),
    (NFL_Merged_df["Pos"] == "SS"),
    (NFL_Merged_df["Pos"] == "K"),
    (NFL_Merged_df["Pos"] == "LS"),
    (NFL_Merged_df["Pos"] == "P")
]
# Create a list of the values to assign for each condition
values = ["Offense", "Offense", "Offense", "Offense", "Offense", "Offense", "Offense", "Offense", "Offense", "Offense",
          "Defense", "Defense", "Defense", "Defense", "Defense", "Defense", "Defense", "Defense", "Defense","Special Teams",
          "Special Teams","Special Teams"]

# create a new column and use np.select to assign values for winning column
NFL_Merged_df["Cat_POS"] = np.select(Positions, values)

# display updated DataFrame
NFL_Merged_df
```

```python
column_names= ["Player","Year","Pos","Cat_POS","Team","BaseSalary","SigningBonus","Cap Hit","Salary%","Conf","Division","Wins","Losses","Ties","Win %","Div_Rank","Scored","Allowed","Net","SOV","SOS","Playoff","Won Superbowl"]
NFL_Merged_df= NFL_Merged_df.reindex(columns=column_names)
NFL_Merged_df
```
