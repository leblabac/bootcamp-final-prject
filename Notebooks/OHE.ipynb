```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder 
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from pathlib import Path
from collections import Counter
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score,mean_squared_error
```

```python
# Load the final merged dataset
OHE_to_load = "../Resources/NFL_Merged_df.csv"
OHE_df = pd.read_csv(OHE_to_load)
OHE_df
```

```python

# drop unneeded columns
# OHE_df = OHE_df.drop(columns=["Player","Wins","Losses","Ties"])
# OHE_df
OHE_df_limited = OHE_df.drop(columns=["Player","Wins","Losses","Ties","Age","Pos","Cat_POS","Team","Conf","Division","Div_Rank","Scored","Allowed","Net","SOV","SOS"])
OHE_df_limited
```

```python
# Using get_dummies and a for loop one hot encode each categorical column and delete the original data.

categorical_columns = ['Pos','Cat_POS', 'Team','Conf', 'Division']
for column in categorical_columns:
    tempdf = pd.get_dummies(OHE_df[column], prefix=column)
    OHE_df = pd.merge(
        left=OHE_df,
        right=tempdf,
        left_index=True,
        right_index=True,
    )
    OHE_df = OHE_df.drop(columns=column)
OHE_df
```

```python
# write df to new csv removing the index

# NFL_OHE_df_path = "../Resources/NFL_OHE_df.csv"
# OHE_df.to_csv(NFL_OHE_df_path, index = False)
OHE_df_limited_path = "../Resources/OHE_df_limited.csv"
OHE_df_limited.to_csv(OHE_df_limited_path, index = False)
```

```python
#set up the X and y values

# X = OHE_df.drop(columns=["Playoff","Won_Superbowl","Win%"])
# y = OHE_df[["Playoff"]]
# y2 = OHE_df[["Win%"]]
# y3 = OHE_df[["Won_Superbowl"]]
# # print(X)
# # print(y3)

X = OHE_df_limited.drop(columns=["Playoff","Won_Superbowl","Win%"])
y = OHE_df_limited[["Playoff"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Creating the scaler instance
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler()
```

```python
# Fitting the scaler
X_scaler = data_scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
# NFL_TTS = pd.DataFrame(NFL_TTS,columns=NFL_Scale.columns)
# NFL_TTS
```

```python
# only for y categorical target
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
LE = label_encoder.fit(y_train)
y_train_encoded = LE.transform(y_train)
y_test_encoded = LE.transform(y_test)
print(y_train_encoded)
```

```python
# Playoff as the target/output/solving for
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78)
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train_encoded)
# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)
print(f"Playofs as Target- Random forest predictive accuracy: {round(accuracy_score(y_test_encoded,y_pred),3)}")
```

```python
# Won_Superbowl as the target/output/solving for

# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78)
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train_encoded)
# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)
print(f"Won_Superbowl target- Random forest predictive accuracy: {round(accuracy_score(y_test_encoded,y_pred),3)}")
```

```python
# Win% as the target/output/solving for errors
# Create a random forest Regressor.
rf_model = RandomForestRegressor(n_estimators=128, random_state=78)

# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
# Evaluate the model
y_pred = rf_model.predict(X_test_scaled)

print(f" Win% Random forest regressor MSE: {round(mean_squared_error(y_test,y_pred),3)}")
```
