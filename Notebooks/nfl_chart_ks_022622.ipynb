```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import our data into pandas from CSV
nfl_file_path = '../Resources/NFL_Merged_df_for_js_chart.csv'
nfl_file_df = pd.read_csv(nfl_file_path)

nfl_file_df
```

```python
# csv file
# Change Salary % heading
# sort by year (Z to A), team (A to Z)
```

```python
# Create temp dataframe based on users selection 
# Year 2021, team MIN

year_value = [2021]
temp_df = pd.DataFrame(nfl_file_df[nfl_file_df.Year.isin(year_value)==True])
```

```python
# team_value = ["MIN"]
temp_df = pd.DataFrame(temp_df[temp_df.Team == "MIN"])
temp_df
```

```python
# Define chart fields, perform data normalization (calculate mean normalization)

x = temp_df.loc[:, ['Salary_Per']]
temp_df['Salary_Per_z'] = (x - x.mean())/x.std()
temp_df['colors'] = ['red' if x < 0 else 'green' for x in temp_df['Salary_Per_z']]
temp_df.sort_values('Salary_Per_z', inplace=True)
temp_df.reset_index(inplace=True)

```

```python
# Draw plot
plt.figure(figsize=(20, 20), dpi=80)
plt.hlines(y=temp_df.index, xmin=0, xmax=temp_df.Salary_Per_z)
for x, y, tex in zip(temp_df.Salary_Per_z, temp_df.index, temp_df.Salary_Per_z):
        t = plt.text(x, y, round(tex, 2), horizontalalignment="right" if x < 0 else "left",
                     verticalalignment='center', fontdict={"color": "red" if x < 0 else "green", "size": 14})

# formatting
plt.yticks(temp_df.index, temp_df.Player, fontsize=14)
plt.title("Player Salary Percentage Over (Under) Average", fontdict={'size': 20})
plt.grid(linestyle= '--', alpha=0.5)
plt.xlim(-1.5, 7.5)
plt.show()
```
