```python
# Import Dependencies

import pandas as pd
import numpy as np
```

## Player Financials


```python
# load and read csv file into DataFrame

financials_to_load = "NFL_Yearly_Breakdown_combined.csv"
financials_df = pd.read_csv(financials_to_load)
financials_df
```

```python
# Look for missing values

print(financials_df.count())
financials_df.isnull()
```

```python
# Drop columns

financials_df = financials_df.drop(columns=["Rank","Unnamed: 19", "FA Year", 
        "Sign Year", "Sign Age","Cash","Draft Year","AAV","CONT YR","CONT VALUE","Earnings"])
financials_df
```

```python
# Add column for Catagory of position (offence, defence, special teams)

Positions = [
    (financials_df["Pos"] == "C"),
    (financials_df["Pos"] == "FB"),
    (financials_df["Pos"] == "G"),
    (financials_df["Pos"] == "LT"),
    (financials_df["Pos"] == "QB"),
    (financials_df["Pos"] == "RB"),
    (financials_df["Pos"] == "RT"),
    (financials_df["Pos"] == "T"),
    (financials_df["Pos"] == "TE"),
    (financials_df["Pos"] == "WR"),
    (financials_df["Pos"] == "CB"),
    (financials_df["Pos"] == "DE"),
    (financials_df["Pos"] == "DT"),
    (financials_df["Pos"] == "FS"),
    (financials_df["Pos"] == "ILB"),
    (financials_df["Pos"] == "LB"),
    (financials_df["Pos"] == "OLB"),
    (financials_df["Pos"] == "S"),
    (financials_df["Pos"] == "SS"),
    (financials_df["Pos"] == "K"),
    (financials_df["Pos"] == "LS"),
    (financials_df["Pos"] == "P")
]
# Create a list of the values to assign for each condition
values = ["Offense", "Offense", "Offense", "Offense", "Offense", "Offense", "Offense", "Offense", "Offense", "Offense",
          "Defense", "Defense", "Defense", "Defense", "Defense", "Defense", "Defense", "Defense", "Defense","Special Teams",
          "Special Teams","Special Teams"]

# create a new column and use np.select to assign values for winning column
financials_df["Cat_POS"] = np.select(Positions, values)

# display updated DataFrame
financials_df.head(32)
```

```python
# Drop rows with future years

values = [2022, 2023, 2024]
financials_df = financials_df[financials_df.Year.isin(values)==False]
financials_df
```

```python
# Check again for missing values

print(financials_df.count())
financials_df.isnull()
```

```python
# Combine the records of renamed teams; look at random selection

financials_df["Team"].replace({"LAR":"LAR/STL", "STL":"LAR/STL", "JAC":"JAX","LV": "LV/OAK","OAK": "LV/OAK", "SD":"LAC/SD", "LAC":"LAC/SD"},inplace=True)
financials_df.sample(25)
```

```python
# Check datatypes
financials_df.info()
```

```python
# changed datatype for 3 columns to integers (base salary, signing bonus, cap hit)

financials_df["BaseSalary"] = financials_df["BaseSalary"].str.replace(',', '')
financials_df["BaseSalary"] = financials_df["BaseSalary"].str.replace('$', '')
financials_df["BaseSalary"] = financials_df["BaseSalary"].astype(int)

financials_df["SigningBonus"] = financials_df["SigningBonus"].str.replace(',', '')
financials_df["SigningBonus"] = financials_df["SigningBonus"].str.replace('$', '')
financials_df["SigningBonus"] = financials_df["SigningBonus"].astype(int)

financials_df["Cap Hit"] = financials_df["Cap Hit"].str.replace(',', '')
financials_df["Cap Hit"] = financials_df["Cap Hit"].str.replace('$', '')
financials_df["Cap Hit"] = financials_df["Cap Hit"].astype(int)


financials_df.info()
financials_df.head()
```

```python
financials_df = financials_df.rename(columns={"Cap Hit":"Cap_Hit"})
financials_df
```

```python
# Export financial Dataframe to csv

financials_df_path = "financials_df.csv"
financials_df.to_csv(financials_df_path, index=False)
```

## Team Standings

```python
# load and read csv file into DataFrame

standings_to_load = "nfl_standings.csv"
nfl_standings_df = pd.read_csv(standings_to_load)
nfl_standings_df
```

```python
# Review dataframe

nfl_standings_df.info()
```

```python
# Change column headings (to be consistent with financial dataframe)

nfl_standings_df = nfl_standings_df.rename(
    columns={"season": "Year", "pct": "Win%","conf":"Conf","division":"Division",
    "team":"Team","wins":"Wins","losses":"Losses","ties":"Ties","div_rank":"Div_Rank",
    "scored":"Scored","allowed":"Allowed","net":"Net","sov":"SOV","sos":"SOS","playoff":"Playoff"})
nfl_standings_df
```

```python
# Combine records of renamed teams

nfl_standings_df["Team"].replace({"LA": "LAR/STL", "STL": "LAR/STL", "LV": "LV/OAK",
                                 "OAK": "LV/OAK", "SD": "LAC/SD", "LAC": "LAC/SD"}, inplace=True),
nfl_standings_df
```

```python
# Drop data (rows) prior to 2011,

nfl_standings_df.drop(
    nfl_standings_df.loc[nfl_standings_df["Year"] < 2011].index, inplace=True)
nfl_standings_df
```

```python
# Add column for superbowl champions

# create conditions for superbowl win for teams that made playoffs
superbowl_conditions = [
    (nfl_standings_df["Playoff"] == "WonSB"),
    (nfl_standings_df["Playoff"] == "LostSB"),
    (nfl_standings_df["Playoff"] == "LostCC"),
    (nfl_standings_df["Playoff"] == "LostDV"),
    (nfl_standings_df["Playoff"] == "LostWC")
]

# Create a list of the values to assign for each condition
values = ["Yes", "No", "No", "No", "No"]

# create a new column and use np.select to assign values for winning column
nfl_standings_df["Won_Superbowl"] = np.select(superbowl_conditions, values)

# replace null values with No for teams that did not make playoffs
nfl_standings_df["Won_Superbowl"].replace({"0": "No"}, inplace=True)

# display 1 year of data to ensure 1 champion
nfl_standings_df.head(32)
```

```python
# change playoff column to Yes or No

nfl_standings_df["Playoff"].replace({"LostSB": "Yes", "LostCC": "Yes", "LostDV": "Yes",
                                     "LostWC": "Yes", "WonSB": "Yes"}, inplace=True)

nfl_standings_df["Playoff"].fillna("No", inplace=True)

nfl_standings_df
```

```python
# Drop seed column

nfl_standings_df = nfl_standings_df.drop(columns="seed")
nfl_standings_df
```

```python
# Export team standings Dataframe to csv

nfl_standings_df_path = "standings_df.csv"
nfl_standings_df.to_csv(nfl_standings_df_path, index=False)
nfl_standings_df
```

## Combine Datasets

```python
# Merge two Dataframes

NFL_Merged_df = pd.merge(financials_df, nfl_standings_df,  how='left', 
                left_on=['Year','Team'], right_on = ['Year','Team'])
NFL_Merged_df
```

```python
# Confirm all columns have merged

NFL_Merged_df.columns
```

```python
# Change column order

column_names= ["Player","Age","Year","Pos","Cat_POS","Team","BaseSalary","SigningBonus","Cap_Hit","Salary%","Conf","Division","Wins","Losses","Ties","Win%","Div_Rank","Scored","Allowed","Net","SOV","SOS","Playoff","Won_Superbowl"]
NFL_Merged_df= NFL_Merged_df.reindex(columns=column_names)
NFL_Merged_df
```

```python
# Export merged dataframe to csv

NFL_Merged_df_path = "NFL_Merged_df.csv"
NFL_Merged_df.to_csv(NFL_Merged_df_path, index=False)
NFL_Merged_df
```
